```python
import numpy as np
from matplotlib import pyplot as plt
import jax.numpy as jnp
from jax import jit, grad, jacfwd, jacrev
```

```python
Q = np.diag([0.5, 1.0])

# Only for contour plot
def plot_f(X, Y):
    # equivalent to 0.5 * ([X,Y] - [1,0]).T @ Q @ ([X,Y] - [1,0])
    return 0.5*(0.5*(X-1)**2 + Y**2)

def f(x):
    B = np.array([1, 0])
    return 0.5 * (x - B).T @ Q @ (x - B)

def grad_f(x):
    B = np.array([1, 0])
    return Q @ (x - B)

def hess_f(x):
    return Q
```

```python
def c(x):
    return x[0]**2 + 2*x[0] - x[1]

# Caution, output of ∂c should be a vector for gradient calculation 
def partial_c(x):
    return np.array([[2 * x[0] + 2, -1]])

def partial_c_jnp(x):
    return jnp.array([2 * x[0] + 2, -1])
```

## Visualization

```python
def plot_landscape():

    x = np.linspace(-4, 4, 100)
    y = np.linspace(-4, 4, 100)
    X, Y = np.meshgrid(x, y)
    Z = plot_f(X, Y)
    contour = plt.contour(X, Y, Z)
    
    contour = plt.contour(X, Y, Z)
    xc = np.linspace(-3.2, 1.2, 100)
    plt.plot(xc, xc**2 + 2.0*xc, color='blue')
    
plot_landscape()
plt.show()
```

```python
def gauss_newton_step(x0, λ0, verbose=False):
    
    h_L = hess_f(x0)
    C = partial_c(x0)
    d_L = grad_f(x0) + C * λ0
    
    KKT_lhs = np.block([ [h_L, C.T], [C, np.zeros((1,1))] ]) 
    KKT_rhs = np.block([ -d_L, np.array([[-c(x0)]]) ]).T
    
    d_z = np.linalg.inv(KKT_lhs).dot(KKT_rhs)
    d_x = d_z[0:2].flatten()
    d_λ = d_z[2].flatten()
    # d_x = d_z[0:2].flatten().tolist()
    # d_λ = d_z[2].flatten().tolist()

    return d_x, d_λ
```

## Normal Case with Overshooting

```python
x_guess = np.array([[-1.0, -1.0]])
λ_guess = np.array([[0.0]])

plot_landscape()
plt.scatter(x_guess[0][0], x_guess[0][1])
plt.show()
```

```python
dx, dλ = gauss_newton_step(x_guess[-1], λ_guess[-1])
newx, newλ = x_guess[-1] + dx, λ_guess[-1] + dλ
x_guess = np.append(x_guess, [newx], axis=0)
λ_guess = np.append(λ_guess, [newλ], axis=0)

plot_landscape()
plt.scatter(x_guess[:,0], x_guess[:,1])
plt.show()
```

# Merit Function

## Function 1. Quadratic Form

```python
@jit
def KKT_hessian(x, λ0):
    return jnp.array(grad_f(x)) + jnp.array(partial_c_jnp(x)) * λ0
```

```python
def P(x, λ):
    C = partial_c(x)
    dx_L = grad_f(x) + C * λ
    d_L = -1 * np.block([ dx_L, c(x)]).flatten()
    # print(f"d_L: {d_L}")
    # print(f"output: {np.dot(d_L, d_L)}")
    return 0.5 * np.dot(d_L, d_L)

def d_P(x, λ):
    C = partial_c(x)
    dx_L = grad_f(x) + C * λ
    d_L = -1 * np.block([ dx_L, c(x)]).flatten()
    H, _ = jacrev(KKT_hessian, argnums=(0,1))(jnp.array(x), jnp.array(λ))
    
    h_L = np.block([ [H, C.T], [C, np.zeros((1,1))] ]) 
    return h_L @ d_L
```

```python
x_guess = np.array([[-1.0, -1.0]])
λ_guess = np.array([[0.0]])

plot_landscape()
plt.scatter(x_guess[0][0], x_guess[0][1])
plt.show()
```

```python
dx, dλ = gauss_newton_step(x_guess[-1], λ_guess[-1])

α = 1.0
θ = 0.5
ρ = 0.01

while P(x_guess[-1]+α*dx, λ_guess[-1]+α*dλ) > P(x_guess[-1], λ_guess[-1]) + ρ*α*np.dot(d_P(x_guess[-1], λ_guess[-1]), np.append(dx, dλ)):
    α = α * θ

newx, newλ = x_guess[-1] + α*dx, λ_guess[-1] + α*dλ
x_guess = np.append(x_guess, [newx], axis=0)
λ_guess = np.append(λ_guess, [newλ], axis=0)

plot_landscape()
plt.scatter(x_guess[:,0], x_guess[:,1])
plt.show()
```

## Function 1. Quadratic Form

```python
p = 1.0

def P(x, λ):
    # print(f"c(x): {c(x)}")
    return f(x) + p * c(x)

def d_P(x, λ):
    C = partial_c(x)
    b1 = grad_f(x) + p * C * np.sign(c(x))
    b2 = np.zeros(len(λ))
    return np.block([ b1[0], b2 ])
```

```python
x_guess = np.array([[-1.0, -1.0]])
λ_guess = np.array([[0.0]])

plot_landscape()
plt.scatter(x_guess[0][0], x_guess[0][1])
plt.show()
```

```python
dx, dλ = gauss_newton_step(x_guess[-1], λ_guess[-1])

α = 1.0
θ = 0.5
ρ = 0.01

while P(x_guess[-1]+α*dx, λ_guess[-1]+α*dλ) > P(x_guess[-1], λ_guess[-1]) + ρ*α*np.dot(d_P(x_guess[-1], λ_guess[-1]), np.append(dx, dλ)):
    α = α * θ

newx, newλ = x_guess[-1] + α*dx, λ_guess[-1] + α*dλ
x_guess = np.append(x_guess, [newx], axis=0)
λ_guess = np.append(λ_guess, [newλ], axis=0)

plot_landscape()
plt.scatter(x_guess[:,0], x_guess[:,1])
plt.show()
```

## Function 3. Augmented Lagrangian

```python
p = 1.0

def P(x, λ):
    # print(f"c(x): {c(x)}")
    return f(x) + λ * c(x) + 0.5 * p * c(x) * c(x)

def d_P(x, λ):
    C = partial_c(x)
    b1 = grad_f(x) + λ * C + p * c(x) * C
    b2 = c(x)
    return np.block([ b1[0], b2 ])
```

```python
x_guess = np.array([[-1.0, -1.0]])
λ_guess = np.array([[0.0]])

plot_landscape()
plt.scatter(x_guess[0][0], x_guess[0][1])
plt.show()
```

```python
dx, dλ = gauss_newton_step(x_guess[-1], λ_guess[-1])

α = 1.0
θ = 0.5
ρ = 0.01

while P(x_guess[-1]+α*dx, λ_guess[-1]+α*dλ) > P(x_guess[-1], λ_guess[-1]) + ρ*α*np.dot(d_P(x_guess[-1], λ_guess[-1]), np.append(dx, dλ)):
    α = α * θ

newx, newλ = x_guess[-1] + α*dx, λ_guess[-1] + α*dλ
x_guess = np.append(x_guess, [newx], axis=0)
λ_guess = np.append(λ_guess, [newλ], axis=0)

plot_landscape()
plt.scatter(x_guess[:,0], x_guess[:,1])
plt.show()
```
