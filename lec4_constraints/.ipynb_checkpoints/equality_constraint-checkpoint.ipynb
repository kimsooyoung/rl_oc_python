```python
import numpy as np
from matplotlib import pyplot as plt
import jax.numpy as jnp
from jax import jit, grad, jacfwd, jacrev
```

```python
Q = np.diag([1.0, 1.0])

# Only for contour plot
def plot_f(X, Y):
    # equivalent to 0.5 * ([X,Y] - [1,0]).T @ Q @ ([X,Y] - [1,0])
    return 0.5*(0.5*(X-1)**2 + Y**2)

def f(x):
    B = np.array([1, 0])
    return 0.5 * (x - B).T @ Q @ (x - B)

def grad_f(x):
    B = np.array([1, 0])
    return Q @ (x - B)

def hess_f(x):
    return Q
```

```python
def c(x):
    return x[0]**2 + 2*x[0] - x[1]

# Caution, output of âˆ‚c should be a vector for gradient calculation 
def partial_c(x):
    return np.array([[2 * x[0] + 2, -1]])

def partial_c_jnp(x):
    return jnp.array([2 * x[0] + 2, -1])
```

## Visualization

```python
def plot_landscape():

    x = np.linspace(-4, 4, 100)
    y = np.linspace(-4, 4, 100)
    X, Y = np.meshgrid(x, y)
    Z = plot_f(X, Y)
    contour = plt.contour(X, Y, Z)
    
    xc = np.linspace(-3.2, 1.2, 100)
    plt.plot(xc, xc**2 + 2.0*xc, color='blue')
    
plot_landscape()
plt.show()
```

## Full Newton Method

```python
@jit
def KKT_hessian(x, Î»0):
    return jnp.array(grad_f(x)) + jnp.array(partial_c_jnp(x)) * Î»0
```

```python
def newton_step(x0, Î»0, verbose=False):
    
    C = partial_c(x0)
    
    d_xL = grad_f(x0) + C * Î»0
    d_Î»L = c(x0)

    h_xL, _ = jacrev(KKT_hessian, argnums=(0,1))(jnp.array(x0), jnp.array(Î»0))

    KKT_lhs = np.block([ [h_xL, C.T], [C, np.zeros((1,1))] ]) 

    KKT_rhs = np.block([ -d_xL, np.array([[-c(x0)]]) ]).T
    
    d_z = np.linalg.inv(KKT_lhs).dot(KKT_rhs)
    d_x = d_z[0:2].flatten().tolist()
    d_Î» = d_z[2].flatten().tolist()

    if verbose is True:
        print(f"C: {C} / {C.shape}")
        print(f"d_xL: {d_xL}, {d_xL.shape} / d_Î»L: {d_Î»L}")
        print(f"x0: {x0} / Î»0: {Î»0}")
        print(f"h_xL: {h_xL}")
        print(f"KKT_lhs: {KKT_lhs}")
        print(f"KKT_rhs: {KKT_rhs}")
        print(f"d_z: {d_z}")
        print(f"d_x: {d_x}")
        print(f"d_Î»: {d_Î»}")

    return x0 + d_x, Î»0 + d_Î»
```

## Normal Converge Case

```python
x_guess = np.array([[-1.0, -1.0]])
Î»_guess = np.array([[0.0]])
plot_landscape()
plt.scatter(x_guess[0][0], x_guess[0][1])
plt.show()
```

```python
newx, newÎ» = newton_step(x_guess[-1], Î»_guess[-1])
x_guess = np.append(x_guess, [newx], axis=0)
Î»_guess = np.append(Î»_guess, [newÎ»], axis=0)

plot_landscape()
plt.scatter(x_guess[:,0], x_guess[:,1])
plt.show()
```

## Exception Case (Hessian Not Posdef) 

```python
x_guess = np.array([[-3.0, 2.0]])
Î»_guess = np.array([[0.0]])
plot_landscape()
plt.scatter(x_guess[0][0], x_guess[0][1])
plt.show()
```

```python
newx, newÎ» = newton_step(x_guess[-1], Î»_guess[-1])
x_guess = np.append(x_guess, [newx], axis=0)
Î»_guess = np.append(Î»_guess, [newÎ»], axis=0)

plot_landscape()
plt.scatter(x_guess[:,0], x_guess[:,1])
plt.show()
```

### Let's Check Hessian of L

```python
x0 = np.array([-3.0, 2.0])
Î»0 = np.array([0.0])

Hessian_0, _ = jacrev(KKT_hessian, argnums=(0,1))(jnp.array(x0), jnp.array(Î»0))
Hessian_end, _ = jacrev(KKT_hessian, argnums=(0,1))(x_guess[-1], Î»_guess[-1])

Hessian_0, Hessian_end
```

## Gauss-Newton Method

```python
def gauss_newton_step(x0, Î»0, verbose=False):
    
    C = partial_c(x0)
    
    d_xL = grad_f(x0) + C * Î»0
    d_Î»L = c(x0)

    # h_xL, _ = jacrev(KKT_hessian, argnums=(0,1))(jnp.array(x0), jnp.array(Î»0))
    h_xL = hess_f(x0)

    KKT_lhs = np.block([ [h_xL, C.T], [C, np.zeros((1,1))] ]) 
    KKT_rhs = np.block([ -d_xL, np.array([[-c(x0)]]) ]).T
    
    d_z = np.linalg.inv(KKT_lhs).dot(KKT_rhs)
    d_x = d_z[0:2].flatten().tolist()
    d_Î» = d_z[2].flatten().tolist()

    if verbose is True:
        print(f"C: {C} / {C.shape}")
        print(f"d_xL: {d_xL}, {d_xL.shape} / d_Î»L: {d_Î»L}")
        print(f"x0: {x0} / Î»0: {Î»0}")
        print(f"h_xL: {h_xL}")
        print(f"KKT_lhs: {KKT_lhs}")
        print(f"KKT_rhs: {KKT_rhs}")
        print(f"d_z: {d_z}")
        print(f"d_x: {d_x}")
        print(f"d_Î»: {d_Î»}")

    return x0 + d_x, Î»0 + d_Î»
```

```python
x_guess = np.array([[-3.0, 2.0]])
Î»_guess = np.array([[0.0]])
plot_landscape()
plt.scatter(x_guess[0][0], x_guess[0][1])
plt.show()
```

```python
newx, newÎ» = gauss_newton_step(x_guess[-1], Î»_guess[-1], False)
x_guess = np.append(x_guess, [newx], axis=0)
Î»_guess = np.append(Î»_guess, [newÎ»], axis=0)

plot_landscape()
plt.scatter(x_guess[:,0], x_guess[:,1])
plt.show()
```
