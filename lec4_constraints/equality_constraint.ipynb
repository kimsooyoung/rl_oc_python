```python
Q = np.diag([1, 1])

# Only for contour plot
def plot_f(X, Y):
    # equivalent to 0.5 * ([X,Y] - [1,0]).T @ Q @ ([X,Y] - [1,0])
    return 0.5*(0.5*(X-1)**2 + Y**2)

def f(x):
    B = np.array([1, 0])
    return 0.5 * (x - B).T @ Q @ (x - B)

def grad_f(x):
    B = np.array([1, 0])
    return Q @ (x - B)

def hess_f(x):
    return Q
```

```python
def plot_landscape():

    x = np.linspace(-4, 4, 100)
    y = np.linspace(-4, 4, 100)
    X, Y = np.meshgrid(x, y)
    Z = plot_f(X, Y)
    contour = plt.contour(X, Y, Z)
    
    contour = plt.contour(X, Y, Z)
    xc = np.linspace(-3.2, 1.2, 100)
    plt.plot(xc, xc**2 + 2.0*xc)
    plt.show()
    
plot_landscape()
```
